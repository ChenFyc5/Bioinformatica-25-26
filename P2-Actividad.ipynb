#### 7.- Con la secuencia y la lista siguiente, escribe un algoritmo que cuente el número de veces que aparece un elemento de la lista en la secuencia. Pista: hacer un diccionario para almacenar los resultados puede ser buena idea

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python
def count_dinucleotides(sequence, dinucleotide_list):
    counts = {dino: 0 for dino in dinucleotide_list}

    for i in range(len(sequence) - 1):
        current_dinucleotide = sequence[i:i+2]
        if current_dinucleotide in counts:
            counts[current_dinucleotide] += 1
    return counts

results = count_dinucleotides(dna, dinucleotides)
for dinucleotide, count in results.items():
    print(f"El dinucleótido '{dinucleotide}' aparece {count} veces.")
```

##### ¿Cuántos counts tiene TC?

```python
tc_counts = results.get('TC', 0)
print(f"El dinucleótido 'TC' tiene {tc_counts} counts.")
```

##### ¿Puedes imprimir los dinuleótidos que tengan counts igual a 2?

```python
print("Dinucleótidos con counts igual a 2:")
for dinucleotide, count in results.items():
    if count == 2:
        print(f"El dinucleótido '{dinucleotide}' aparece {count} veces.")
```
