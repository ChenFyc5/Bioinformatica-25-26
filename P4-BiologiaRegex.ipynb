##### 2.- Crea dos archivos. Uno almacena los accession que empiezen por 'a' y el otro el resto

```python
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']
```

```python
accessions_a = [acc for acc in accs if acc.startswith('a')]
accessions_resto = [acc for acc in accs if not acc.startswith('a')]

with open('accessions_a.txt', 'w') as f:
    f.write('\n'.join(accessions_a))

with open('accessions_resto.txt', 'w') as f:
    f.write('\n'.join(accessions_resto))

print("Resultados:")
print("accessions_a.txt:", accessions_a)
print("accessions_resto.txt:", accessions_resto)
```

##### ¿Qué pasa si no ponemos un argumento? ¿Sabrías hacer una función encapsulada?

```python
def encapsulada(secuencia="ATG", decimales=2):
    def calcular_porcentaje(secuencia, dec):
        secuencia = secuencia.lower()
        total = len(secuencia)
        if total == 0:
            return 0.0
        numero_a = secuencia.count('a')
        return round((numero_a / total) * 100, dec)
    
    return calcular_porcentaje(secuencia, decimales)
```
