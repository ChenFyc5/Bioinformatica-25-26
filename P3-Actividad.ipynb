## JUEGO DE ADIVINANZAS

#### Escribe un programa que coja un número entero del 1 al 100, y los jugadores tienen que tratar de adivinar el número.

#### Las reglas son las siguientes:

1. Si un jugador escoje un número menor que 1 o mayor que 100, el programa dice "FUERA DE LÍMITES"
2. En el primer turno del jugador, si está en un rango de 10 del número dice "¡CALIENTE!". Si está fuera de ese rango dice "¡FRIO!"
3. En todas las rondas siguientes, si el jugador se acerca más que la vez anterior, dice "¡MAS CALIENTE!". Si está más lejos que la anterior vez dice "¡MAS FRIO!"
4. Cuando el jugador acierte el número, les dirá que han acertado y cuántos intentos les ha llevado





##### PISTAS

2. Haz una introducción de tu juego y explica las reglas

5. Modifica ese bucle para que compare nuestro número con el intento. Si se acierta usa break, si no puedes usar continue y tienes que seguir preguntando al jugador por nuevos intentos

```python
def reglas():
    print("Bienvenido a ADIVINA EL NUMERO!")
    print("He pensado un número entre 1 y 100.")
    print("Las reglas son las siguientes:")
    print("1. Si escoges un número fuera de 1 a 100, el programa dirá 'FUERA DE LÍMITES'.")
    print("2. En tu primer intento: si estás a 10 de distancia del número secreto, diré '¡CALIENTE!', si no, '¡FRÍO!'.")
    print("3. En los intentos siguientes: si te acercas más que antes, diré '¡MÁS CALIENTE!', si te alejas, '¡MÁS FRÍO!'.")
    print("4. Cuando aciertes, te diré cuántos intentos necesitaste. MUCHA SUERTE!\n")
```

```python
def jugar():
    import random

    numero_secreto = random.randint(1, 101)
    intentos = 0
    intento_anterior = None 
    diferencia_anterior = None
    
    while True:
        try:
            intento = int(input("Elige un número entre 1 y 100: "))
            if intento < 1 or intento > 100:
                print("El número tiene que estar entre 1 y 100")
                intentos -= 1
                continue
            intentos += 1
            if intento == numero_secreto:
                print(f"¡Has acertado! Has adivinado el número en {intentos} intentos.")
                break
            diferencia_actual = abs(numero_secreto - intento)
            if intentos == 1:
                if diferencia_actual <= 10:
                    print("¡CALIENTE!")
                else:
                    print("¡FRIO!")
            else:
                if diferencia_actual < diferencia_anterior:
                    print("¡MAS CALIENTE!")
                elif diferencia_actual > diferencia_anterior:
                    print("¡MAS FRIO!")
                else:
                    print ("¡Es el mismo número!")
            intento_anterior = intento
            diferencia_anterior = diferencia_actual
    
        except ValueError:
            print("Por favor, introduce un número valido.")
```

```python
reglas()
```

```python
jugar()
```
